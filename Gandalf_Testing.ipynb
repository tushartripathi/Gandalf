```python

def get_average_percentage_change(ticker):

    today = datetime.today()
    start_date = today - timedelta(days=365)

    # Download data for the last 365 days
    data = yf.download(ticker, start=start_date.strftime('%Y-%m-%d'), end=today.strftime('%Y-%m-%d'), interval="1d")

    # print(data)
    if data.empty:
        raise ValueError(f"No data available for ticker: {ticker}")

    if isinstance(data.columns, pd.MultiIndex):
        data.columns = [' '.join(col).strip() for col in data.columns]
        # print("Flattened columns:", data.columns)

    low_col = next((col for col in data.columns if 'Low' in col), None)
    high_col = next((col for col in data.columns if 'High' in col), None)

    if not low_col or not high_col:
        raise ValueError(f"Missing 'Low' or 'High' columns in data for ticker: {ticker}")

    df = pd.DataFrame({
        'Low': data[low_col],
        'High': data[high_col]
    })

    # print(df)
    df['Percentage Change'] = ((df['High'] - df['Low']) / df['Low']) * 100

    average_percentage_change = df['Percentage Change'].mean()

    # print(f"Average Percentage Change for {ticker}: {average_percentage_change:.2f}%")

    results.append({'Ticker': ticker, 'Average Percentage Change': average_percentage_change})

# get_average_percentage_change('RELIANCE.NS', results)
# print(result)
# print("Results list:", results)
```

This functin is the heart of the code.

Calling this will get the prices of stock + calculate mean percentage + add values to the results param

```python
def callGandalof():

  results.clear()
  for ticker in nse_top_600_ns:
    get_average_percentage_change(ticker)
  print("Printing sorted result")
  results.sort(key=lambda x: x.get('avper', 0), reverse=True)
  for result in results:
      print(result)
```
